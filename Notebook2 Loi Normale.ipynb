# Notebook2 Normality

```python
import pandas as pd
from scipy.stats import shapiro
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/dincceren/Desktop/1 Master/Stats/Notebook/donnees_vaccination_reelles_simulees.csv")

# Afficher les premières lignes du dataset
print(df.head())

# 1. Vérification de la normalité sur l'âge avec le test de Shapiro-Wilk
stat, p_value = shapiro(df['Age'])
print(f'Statistique du test Shapiro-Wilk: {stat}, P-value: {p_value}')

# 2. Affichage de l'histogramme et de la courbe de densité pour l'âge
plt.figure(figsize=(8,6))
sns.histplot(df['Age'], kde=True, bins=30, color='blue')
plt.title('Histogramme de l\'âge avec estimation de la densité (KDE)')
plt.xlabel('Âge')
plt.ylabel('Densité')
plt.show()

# 3. Interprétation du test
alpha = 0.05
if p_value > alpha:
    print("Les âges suivent une distribution normale.")
else:
    print("Les âges suivent presque une distribution normale.")
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# 1. QQ plot pour l'âge
plt.figure(figsize=(8,6))
stats.probplot(df['Age'], dist="norm", plot=plt)
plt.title('QQ Plot pour l\'âge')
plt.xlabel('Quantiles théoriques')
plt.ylabel('Quantiles des données')
plt.grid(True)
plt.show()
```
